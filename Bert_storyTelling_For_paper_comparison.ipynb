```python
def path_finder(start_, all_d):
  r= 0

  #len(all_d[0])-1
  res = list()
  res.append(start_)
  while(r<len(bert_embedding.keys())-1):
    if start_ not in list_of_ts_[r]:
      r +=1
    
    try: # there is no story to follow
      if len(all_d[r][start_].values()) <1:
        r +=1
        continue
    except:
      return res
    min_val = min(all_d[r][start_].values())
    min_ind = [x for x,y in all_d[r][start_].items() if y==min_val]
    res.append(min_ind[0])
    if start_ in list_of_ts_[r]:
      del all_d[r][start_]
      for w in all_d[r].keys():
        del all_d[r][w][start_]
    if r == len(bert_embedding.keys())-1:
      break
    if start_ in list_of_ts_[r+1]:
      del all_d[r+1][start_]
      for w in all_d[r+1].keys():
        del all_d[r+1][w][start_]
    if r == len(bert_embedding.keys())-1:
      break

    start_ = min_ind[0]
  return res
```

```python
dfp = 0
for mu in list_of_ts_:
  print(len(mu), '  ', mu)
  # print(dfp)
  dfp += len(mu)
```

```python
full_story = list()
for ni in list_of_ts_[0]:
  full_story.append(path_finder(ni, copy.deepcopy(bert_embedding)))
```
